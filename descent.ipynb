```python
import numpy as np
from sklearn import datasets

import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt

from descent import line_search, gradient_descent
```

```python
def ss1(x):
    return np.sum(x ** 2, axis=0)

def ss1_grad(x):
    return 2 * x
```

```python
def ss2(x):
    return np.sum(x[0] ** 2 + 5 * x[1] ** 2 + x[0] * x[1], axis=0)

def ss2_grad(x):
    return np.array([2 * x[0] + x[1], 10 * x[1] + x[0]])
```

```python
x = np.linspace(-100, 100)
x0 = (np.random.rand(2) - 0.5) * 200
```

```python
traj1 = np.array(gradient_descent(ss1, ss1_grad, x0, compute_alpha=True, return_trajectory=True))
```

```python
plt.figure(figsize=(8, 8))
plt.plot(*x0, 'o', color='r', markersize=10, label='start')
plt.plot(*traj1.T, '--', linewidth=2, color='r', label='trajectory')
plt.plot(*traj1[-1], 'x', markersize=10, markeredgewidth=3, color='r', label='end')
plt.contourf(x, x, np.array([[ss1(np.array([i, j])) for i in x] for j in x]))
plt.legend();
```

```python
traj2 = np.array(gradient_descent(ss2, ss2_grad, x0, compute_alpha=True, return_trajectory=True))
```

```python
plt.figure(figsize=(8, 8))
plt.plot(*x0, 'o', color='r', markersize=10, label='start')
plt.plot(*traj2.T, '--', linewidth=2, color='r', label='trajectory')
plt.plot(*traj2[-1], 'x', markersize=10, markeredgewidth=3, color='r', label='end')
plt.contourf(x, x, np.array([[ss2(np.array([i, j])) for i in x] for j in x]))
plt.legend();
```
